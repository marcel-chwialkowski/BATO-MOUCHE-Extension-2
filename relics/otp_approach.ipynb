```python
filosofi_file = "../data/Filosofi2015_carreaux_200m_paris.gpkg" 

#get number of amenities in each place
pgdf = hs.get_POI_cat_on_INSPIRE_grid(filosofi_file) 

#add up the number of housing units
pgdf['Log'] = pgdf['Log_inc'] + pgdf['Log_av45'] + pgdf['Log_45_70'] + pgdf['Log_70_90'] + pgdf['Log_ap90']
```

Below, we calculate travel times between squares

```python
pgdf_cp = pgdf.copy()
pgdf_cp['geometry'] = pgdf_cp['geometry'].apply(lambda x : x.centroid)
time_by_id = hs.otp_calculate_time_matrix(pgdf_cp)
#since this takes a ton of time, i'd rather get reliable progress notes, so i guess i will restart?
```

as this calculation takes a long time usually, the results are saved in time_by_id.csv

```python
#time_by_id_fetched = read csv
weights_by_id = hs.transfer_time_to_weight(time_by_id, 30)
weights_by_id.to_csv("data/weights_by_id.csv")
```

```python
interestVar = ['Log','Log_soc','restaurant',
       'culture and art', 'education', 'food_shops', 'fashion_beauty',
       'supply_shops']
accessibility_measures = hs.calculate_2SFCA_accessibility(pgdf,interestVar,weights_by_id)
```
